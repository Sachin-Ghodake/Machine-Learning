```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#import dataset
df=pd.read_csv("book.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

#### 1. Association rules with 10% Support and 70% confidence¶

```python
#with 10% support
frequent_items=apriori(df,min_support=0.1,use_colnames=True)
```

```python
frequent_items
```

```python
#with 70% confidence
rules=association_rules(frequent_items,metric='lift',min_threshold=0.7)
```

```python
rules
```

```python
rules[rules['lift']>1]
```

```python
# A leverage value of 0 indicates independence. Range will be [-1,1]
# A high conviction value means that the consequent is highly depending on the antecedent and range [0,inf]
```

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
rules[['support', 'confidence', 'lift']].hist(figsize = (5,5))
plt.show()
```

#### 2. Association rules with 20% Support and 60% confidence

```python
# With 20% Support
frequent_itemsets2=apriori(df,min_support=0.20,use_colnames=True)
frequent_itemsets2
```

```python
# With 60% confidence
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.6)
rules2
```

```python
rules2[['support', 'confidence', 'lift']].hist(figsize=(5,5))
plt.show()
```
