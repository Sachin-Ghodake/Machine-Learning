# Model Building :

```python
X = df1[['Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax','Weight']]
Y = df1[['Price']]

model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
```

```python
# Coefficient
model.params
```

```python
# Calculating T-value for each column
print("T-value :")
print(model.tvalues)
```

```python
# Calculating P-value for each column
print("P-values :")
print(model.pvalues)
```

```python
# R-squared & Adjusted R-squared values
print('R-squared value :',model.rsquared)
print('Adjusted R-squared value :',model.rsquared_adj)
```

```python
# Model summary
model.summary()
```

***Solving multicollinearity :***

```python
# Cook's distance

influence = model.get_influence()

summ_df1 = influence.summary_frame()

summ_df1.sort_values('cooks_d', ascending=False)
```

```python
# Influence plot
plt.rcParams['figure.figsize'] = (10,5)
influence.plot_influence()
```

# Calculating Variance Inflation Factor [VIF] :

```python
vif = pd.DataFrame()

vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

vif["Features"] = X.columns

vif.round(1)
```

```python
# Droping high VIF value
# Removing "Weight" column & testing the model

new_X = df1[['Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax']]
```

```python
new_model = sm.OLS(Y, new_X).fit()
new_predictions = new_model.predict(new_X)

new_model.summary()
```

# Final Model :

```python
# Removed Index with highest Cook's distance to remove the hightest influencial ROW

new_df = df1.drop(df1.index[80])
```

```python
new_X = new_df[['Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax']]
new_Y = new_df[['Price']]

final_model = sm.OLS(new_Y, new_X).fit()
predictions = final_model.predict(new_X)

final_model.summary()
```

**Conclusion :**
- p-value < 0.01, Thus the model is accepted.
- coefficient = (-132.01), Thus if the value of x increased by 1, the predicted value of Price will decrease by 132.01
- Adj. R-sqared = 0.983 Thus the model explains 98.3% of the variance in dependent variable.
